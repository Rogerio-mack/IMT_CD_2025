### 2.1. Quais os atributos com maior ganho de informação selecionados e quais são os seus valores?



```python
from sklearn.feature_selection import mutual_info_classif
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

# Separate features (X) and target (y)
X = df.drop(['card', 'client_name'], axis=1)
y = df['card']

# Identify categorical features
categorical_cols = X.select_dtypes(include=['object']).columns

# Apply one-hot encoding to categorical features for mutual information calculation
# We use get_dummies here for simplicity in this specific context,
# not necessarily for downstream model training.
for c in categorical_cols:
    X[c] = LabelEncoder().fit_transform(X[c])

# Calculate mutual information scores
mutual_info = mutual_info_classif(X, y, random_state=42)

# Create a pandas Series for better visualization of scores
mutual_info_series = pd.Series(mutual_info, index=X.columns)

# Sort the scores in descending order
mutual_info_series.sort_values(ascending=False, inplace=True)

# Display the mutual information scores
print("Mutual Information Scores for features:")
display(mutual_info_series)
```

```python
from sklearn.model_selection import train_test_split

predictors = list(mutual_info_series.index[0:4])
X = df[predictors]
y = df['card']

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.25, random_state=42)
X_train.reset_index(drop=True, inplace=True)
X_test.reset_index(drop=True, inplace=True)
y_train.reset_index(drop=True, inplace=True)
y_test.reset_index(drop=True, inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaled_data = scaler.fit_transform(X_train)
X_train = pd.DataFrame(scaled_data, columns=X_train.columns)

scaled_data = scaler.transform(X_test)
X_test = pd.DataFrame(scaled_data, columns=X_test.columns)

display(X_train.head())
display(X_test.head())
```

### 2.2. Qual a acuracidade de cada um dos modelos obtidos e qual o **melhor modelo** (critério de acuracidade)?



### 2.3. No **pior modelo** (critério de acuracidade), qual valor de classe apresenta mais Falsos Positivos, e qual é essa quantidade?

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, f1_score, classification_report, confusion_matrix

# Initialize the models
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'KNN k=6': KNeighborsClassifier(n_neighbors=6),
    'Decision Tree': DecisionTreeClassifier(random_state=42)
}

# Train and evaluate each model
results = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    results[name] = accuracy
    print(f'{name} Accuracy: {accuracy:.4f}')
    confusion = confusion_matrix(y_test, y_pred)
    print("\nMatriz de Confusão:\n", confusion)
    print()
    print(classification_report(y_test, y_pred))

# Optional: Print the best model
best_model_name = max(results, key=results.get)
print(f'\nBest model based on accuracy: {best_model_name}')
```

```python
from sklearn.metrics import classification_report

# Define class names
target_names = ['left', 'right', 'up']

# Generate classification report from the confusion matrix
# Note: classification_report expects true and predicted labels, not the confusion matrix directly.
# To use the confusion matrix, we can reconstruct dummy true and predicted labels
# that would produce this confusion matrix. This is a bit involved.

# A simpler approach is to use the calculated values from the confusion matrix
# to manually calculate precision, recall, and f1-score for each class.
# However, since the user specifically asked for scikit-learn's classification_report,
# let's assume the confusion matrix was derived from some true and predicted labels.

# If you have the original true and predicted labels that generated this cm,
# you would use:
# print(classification_report(true_labels, predicted_labels, target_names=target_names))

# Since we only have the confusion matrix, we can't directly use classification_report.
# Let's provide the classification report based on the manual calculation using the cm values.

# Extract values from the confusion matrix
# Row represents actual class, column represents predicted class
# cm[i, j] is the number of samples of actual class i that were predicted as class j

# Class 'left' (index 0)
tp_left = cm[0, 0]
fn_left = cm[0, 1] + cm[0, 2]
fp_left = cm[1, 0] + cm[2, 0]
tn_left = cm[1, 1] + cm[1, 2] + cm[2, 1] + cm[2, 2] # Sum of all elements not in row 0 or col 0

# Class 'right' (index 1)
tp_right = cm[1, 1]
fn_right = cm[1, 0] + cm[1, 2]
fp_right = cm[0, 1] + cm[2, 1]
tn_right = cm[0, 0] + cm[0, 2] + cm[2, 0] + cm[2, 2]

# Class 'up' (index 2)
tp_up = cm[2, 2]
fn_up = cm[2, 0] + cm[2, 1]
fp_up = cm[0, 2] + cm[1, 2]
tn_up = cm[0, 0] + cm[0, 1] + cm[1, 0] + cm[1, 1]


# Calculate metrics manually
precision_left = tp_left / (tp_left + fp_left) if (tp_left + fp_left) > 0 else 0
recall_left = tp_left / (tp_left + fn_left) if (tp_left + fn_left) > 0 else 0
f1_left = 2 * (precision_left * recall_left) / (precision_left + recall_left) if (precision_left + recall_left) > 0 else 0

precision_right = tp_right / (tp_right + fp_right) if (tp_right + fp_right) > 0 else 0
recall_right = tp_right / (tp_right + fn_right) if (tp_right + fn_right) > 0 else 0
f1_right = 2 * (precision_right * recall_right) / (precision_right + recall_right) if (precision_right + recall_right) > 0 else 0

precision_up = tp_up / (tp_up + fp_up) if (tp_up + fp_up) > 0 else 0
recall_up = tp_up / (tp_up + fn_up) if (tp_up + fn_up) > 0 else 0
f1_up = 2 * (precision_up * recall_up) / (precision_up + recall_up) if (precision_up + recall_up) > 0 else 0

# Calculate overall accuracy
accuracy = np.sum(np.diag(cm)) / np.sum(cm) if np.sum(cm) > 0 else 0

# Print a formatted report similar to classification_report
print("Manual Classification Report:")
print(f"{'':<10s}{'precision':<10s}{'recall':<10s}{'f1-score':<10s}{'support':<10s}")
print("-" * 50)
print(f"{target_names[0]:<10s}{precision_left:<10.2f}{recall_left:<10.2f}{f1_left:<10.2f}{np.sum(cm[0, :]):<10d}")
print(f"{target_names[1]:<10s}{precision_right:<10.2f}{recall_right:<10.2f}{f1_right:<10.2f}{np.sum(cm[1, :]):<10d}")
print(f"{target_names[2]:<10s}{precision_up:<10.2f}{recall_up:<10.2f}{f1_up:<10.2f}{np.sum(cm[2, :]):<10d}")
print("-" * 50)
print(f"{'accuracy':<10s}{'':<10s}{'':<10s}{accuracy:<10.2f}{np.sum(cm):<10d}")

# Note: To get the exact output of scikit-learn's classification_report,
# you would need the original true and predicted labels.
# The manual calculation above provides the same metrics but the formatting
# might differ slightly.
```
