# **Análise de Sentimento de Documentos**

## Introdução
No contexto do processamento de linguagem natural (NLP) a análise de sentimentos, ou mineração de opinião, é análise de documentos para determinar o se o documento expressa um sentimento positivo, negativo ou neutro. Mais recentemente essa análise vem obtida, em geral, com grandes modelos de linguagem (LLMs). Uma aplicação comum é buscar o sentimento de notícias para, por exemplo, obter o sentimento expresso sobre um evento (uma declaração política ou algo relacionado a uma personalidade etc.) para se possa definir uma ação com relação sobre essa receptividade desse evento (uma nova declaração, um esclarecimento ao público pela personalidade etc.). Também vários trabalhos têm sido feitos buscando empregar o sentimento, com relação à notícias do cenário econômico, para a predição de variáveis econômicas e financeiras (preço de ações, índices de preços e inflação etc.).
