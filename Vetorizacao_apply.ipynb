```python
df.apply(lambda x: x.duration/60,axis=1)
```

```python
def score_director(num_critic_for_reviews,director_facebook_likes):

  if director_facebook_likes > 0:
    return num_critic_for_reviews * director_facebook_likes
  else:
    return num_critic_for_reviews
```

```python
df.apply(lambda row: score_director(row.num_critic_for_reviews,row.director_facebook_likes), axis=1)
```

```python
def score_director(row):

  if row.director_facebook_likes > 0:
    return row.num_critic_for_reviews * row.director_facebook_likes
  else:
    return row.num_critic_for_reviews
```

```python
df.apply(lambda row: score_director(row), axis=1)
```

```python
df.apply(score_director,axis=1)
```

# Expressões Regulares

```python
df = pd.read_csv('https://github.com/Rogerio-mack/IMT_CD_2025/raw/main/data/restaurant_inspections2.csv')
df.head()
```

## Selecionar somente os registros com `date` no formato 9999-99-99


```python
df['Date'].str.match("^\d{4}-\d{2}-\d{2}$")
```

### Datas no formato 9999-99-99

```python
df[ df['Date'].str.match("^\d{4}-\d{2}-\d{2}$") ]
```

### Datas NÃO no formato 9999-99-99

```python
df[ ~ df['Date'].str.match("^\d{4}-\d{2}-\d{2}$") ]
```

### Com `import re` e `apply`

```python
import re

def valida_data(data):
  regex = r"^\d{4}-\d{2}-\d{2}$"
  return re.match(regex, data) is not None
```

```python
df['Date'].apply(valida_data)
```

```python
df[ ~ df['Date'].apply(valida_data) ]
```

## Validando datas reais, `date` YYYY-MM-DD, isto é, valores válidos

```python
from datetime import datetime
```

```python
datetime.strptime('1999-12-01','%Y-%m-%d')
```

```python
%%script echo skipping error
datetime.strptime('1999-13-01','%Y-%m-%d')
```

```python
def valida_data(data):
  try:
    datetime.strptime(data,'%Y-%m-%d')
    return True
  except:
    return False
```

```python
df[ ~ df['Date'].apply(valida_data) ]
```
