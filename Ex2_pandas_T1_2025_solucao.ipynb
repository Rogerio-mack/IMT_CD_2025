# Q2.

Considere os dados de `coal_eletricity`. Crie para cada localizaÃ§Ã£o geogrÃ¡fica (`geo`) os valores 'all' e 'mean' com a soma e mÃ©dia de valores de todos os anos como abaixo. Salve o resultado em um dataframe `coal_verti`. Elabore uma soluÃ§Ã£o sem laÃ§os de programa.


````python
coal_horiz = coal_eletricity.pivot_table('coal_electricity_generation_twh', index='geo',columns='year',observed=True)
coal_horiz['mean'] = coal_horiz.mean(axis=1)
coal_horiz['all'] = coal_horiz.drop(columns='mean').sum(axis=1)
coal_verti = coal_horiz.reset_index().melt(id_vars='geo',value_vars=coal_horiz.columns).sort_values('geo').reset_index(drop=True)
coal_verti.columns = coal_eletricity.columns
coal_verti = coal_verti.sort_values(['geo','year']).reset_index(drop=True)

display(coal_verti)
````

````python
#@markdown sanity check
print(coal_verti.shape == (1628, 3))
````

Responda em Q2 com o valor da soma abaixo em:

````python
random.seed(1)
int(coal_verti.loc[random.sample(range(len(coal_verti)), 10)].coal_electricity_generation_twh.sum())
````

# Q4. Pareto 80/20

Considere os dados de `coal_eletricity`. Considere os dados de `coal_eletricity`. **Encontre os maiores geradores em 2019 e determine o pequeno nÃºmero de paÃ­ses responsÃ¡veis por 80% da geraÃ§Ã£o mundial**. Note que os dados possuem valores totais por regiÃ£o que precisam ser excluÃ­dos. VocÃª Ã© livre para elaborar uma soluÃ§Ã£o com ou sem laÃ§os de programa (sim, sem laÃ§o Ã© possÃ­vel).

**Dica.** VocÃª pode querer empregar o comando [`cumsum`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.cumsum.html)


````python
coal_eletricity = coal_eletricity.loc[ ~coal_eletricity.geo.isin([ x for x in coal_eletricity.geo if 'total' in x ])]
coal_80 = coal_eletricity[ coal_eletricity.year == 2019 ].sort_values('coal_electricity_generation_twh',ascending=False)
coal_80 = coal_80.reset_index(drop=True)

coal_80['acum'] = coal_80[['coal_electricity_generation_twh']].cumsum() / coal_80['coal_electricity_generation_twh'].sum()
coal_80[ coal_80['acum'] < 0.82 ]
````

Nesta questÃ£o responda sobre os paÃ­ses maiores geradores de energia que correspondem a 80% da produÃ§Ã£o mundial dos paÃ­ses.

# Q5. Como se relacionam as emissÃµes de CO2 e a geraÃ§Ã£o de energia por carvÃ£o?

Veja o exemplo de join a seguir. E vocÃª pode querer acessar a documentaÃ§Ã£o do `Pandas` para uso do [`merge`](https://pandas.pydata.org/docs/reference/api/pandas.merge.html).





````python
country_entities = pd.read_csv('https://github.com/open-numbers/ddf--gapminder--co2_emission/raw/master/ddf--entities--country.csv')
country_entities.head()
````

````python
co2_emissions = pd.read_csv('https://github.com/open-numbers/ddf--gapminder--co2_emission/raw/master/ddf--datapoints--yearly_co2_emissions_1000_tonnes--by--country--year.csv')
co2_emissions.head()
````

````python
co2_country = country_entities.merge(co2_emissions, how='inner', left_on='country', right_on='country')

# ou simplesmente,
# co2_country = country_entities.merge(co2_emissions)
# jÃ¡ que colunas de mesmo nome sÃ£o automaticamente empregadas no join

co2_country
````

Empregue o exemplo anterior e o dataframe `coal_eletricity` para construir um conjunto de dados `co2_coal` como abaixo.

````python
coal_eletricity = pd.read_csv('https://github.com/open-numbers/ddf--bp--energy/raw/master/ddf--datapoints--coal_electricity_generation_twh--by--geo--year.csv')
coal_eletricity.head()
````

````python
co2_country['name'] = co2_country['name'].str.lower()
co2_coal = co2_country.merge(coal_eletricity, how='inner', left_on=['name','year'], right_on=['geo','year'])

co2_coal = co2_coal.drop(columns=['geo'])
co2_coal
````

````python
co2_coal[ co2_coal.country == 'us']
````

Note que os nomes dos paÃ­ses encontram-se em um formato diferente (lower/upper case) nos dataframes a serem empregados e vocÃª pode querer empregar a funÃ§Ã£o [`str`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.html) para ajustar isso. Responda Q5 com a correlaÃ§Ã£o abaixo.





````python
np.round(co2_coal[['yearly_co2_emissions_1000_tonnes','coal_electricity_generation_twh']].corr().values[0,1],5)
````
