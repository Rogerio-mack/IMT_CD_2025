```python
# prompt: read /content/CD_P2 (respostas) - Respostas ao formulário 1.csv

import pandas as pd

df = pd.read_excel('/content/CD_Ex3_Visualizacao (respostas).xlsx')
print(df.iloc[:,-1])
```

```python
notebooks = df.iloc[:,-1].to_list()
```

```python
# prompt: Faça o download deste único notebook compartilhado no Colab: https://colab.research.google.com/drive/1iKeS0E494vulGfaXEJc2-BtLq4Gr8Wie

# This library is needed to download notebooks from Colab URLs
!pip install nbformat

import pandas as pd
import requests
import nbformat
import os
import json  # Need to import json to save the dictionary content


def download_drive(url, download_dir, filename):
  try:
      # Adjust URL if it's a public sharing link
      if 'colab.research.google.com/drive/' in url:
          # Extract file ID from the URL
          file_id = url.split('/drive/')[1].split('/')[0].split('?')[0]
          print(f"Detected Google Drive file ID: {file_id}")
          # Construct the export URL for the .ipynb format
          # Using drive.google.com/uc?export=download works better for direct download
          download_url = f'https://drive.google.com/uc?export=download&id={file_id}'
          print(f"Attempting download from: {download_url}")
      else:
          # Assume it's a direct link to the raw .ipynb file (less common for shared notebooks)
          download_url = url
          print(f"Attempting download from direct URL: {download_url}")


      response = requests.get(download_url)
      response.raise_for_status()  # Raise an exception for bad status codes

      # Try to parse the content as a notebook to confirm it's a notebook file
      is_notebook = False
      notebook_content = None
      try:
          notebook_content = response.json()
          nbformat.reads(str(notebook_content), as_version=4) # Validate as notebook
          is_notebook = True
          print("Successfully parsed as notebook content.")
      except Exception as e:
          print(f"Warning: The content did not seem to be a valid notebook JSON format. Error: {e}")
          # If it failed to parse as JSON, try saving the content directly,
          # it might be a raw file download that wasn't JSON structured.

      if is_notebook and notebook_content is not None:
          # Save the notebook content to a file
          with open(filename, 'w') as f:
              json.dump(notebook_content, f, indent=2) # Use json.dump to write the dictionary

          print(f"Downloaded notebook from {url} to {filename}")
      elif response.content:
          # If it wasn't JSON, try saving the raw content
          print(f"Attempting to save raw content to {filename}")
          with open(filename, 'wb') as f:
              f.write(response.content)
          print(f"Saved raw content from {url} to {filename}")
      else:
          print(f"No content received from {url}")


  except requests.exceptions.RequestException as e:
      print(f"Error downloading notebook from {url}: {e}")
      print(f"Trying wget as an alternative...")
      # Try wget as an alternative download method
      try:
          !wget -O "{filename}" "{url}"
          print(f"Successfully downloaded using wget to {filename}")
      except Exception as wget_e:
          print(f"Error downloading using wget: {wget_e}")
  except Exception as e:
      print(f"An unexpected error occurred while processing {url}: {e}")

  print("Download process finished.")

  return

#
# teste
#
# Create a directory to store the downloaded notebooks
# download_dir = 'downloaded_notebooks'
# os.makedirs(download_dir, exist_ok=True)

# url = 'https://colab.research.google.com/drive/1iKeS0E494vulGfaXEJc2-BtLq4Gr8Wie'
# filename = os.path.join(download_dir, 'downloaded_notebook.ipynb')

# download_drive(url, download_dir, filename)
```

```python
# Create a directory to store the downloaded notebooks
download_dir = 'downloaded_notebooks'
os.makedirs(download_dir, exist_ok=True)

for i, url in enumerate(notebooks):
    print(f"\n\nProcessing URL at index {i}: {df.iloc[i,2]}")
    filename = os.path.join(download_dir, f'{df.iloc[i,2]}.ipynb')
    print(url)

    if 'github' in url:
      # https://github.com/ViniciusBerti/20241_maua_ecm252_intro_git/blob/main/IMT_CD_P2_2025S1_1.ipynb
      # https://raw.githubusercontent.com/ViniciusBerti/20241_maua_ecm252_intro_git/refs/heads/main/IMT_CD_P2_2025S1_1.ipynb
      url = url.replace('github.com', 'raw.githubusercontent.com').replace('/blob/', '/refs/heads/')
      print(url)
      !wget -O "{filename}" "{url}"

    elif 'https://colab.research.google.com/' in url:

      # print(f"\n\nProcessing URL at index {i}: {df.iloc[i,2]}")
      url = url.split('?usp=sharing')[0]
      download_drive(url, download_dir, filename)

    else:
      print('Neither Git or Colab')
```

```python
# prompt: dada uma pasta com arquivos de notebooks python (.ipynb) faz o print de cada arquivo em um arquivo .pdf

# Montar o Google Drive
from google.colab import drive
# drive.mount('/content/drive')

# Substitua pelo caminho da sua pasta no Drive
notebooks_folder = '/content/downloaded_notebooks'

# Instalar nbconvert e dependências (isso pode falhar ou ser complicado no Colab)
# É muito provável que as dependências LaTeX para PDF não estejam disponíveis facilmente.
# !pip install nbconvert
# !apt update && apt install -y texlive-xetex texlive-fonts-recommended texlive-plain-generic

import os

# Listar arquivos na pasta
notebook_files = [f for f in os.listdir(notebooks_folder) if f.endswith('.ipynb')]

# Tentar converter cada notebook para PDF
for notebook_file in notebook_files:
    notebook_path = os.path.join(notebooks_folder, notebook_file)
    output_filename = os.path.splitext(notebook_file)[0] + '.pdf'
    output_path = os.path.join(notebooks_folder, output_filename) # Salvar PDF na mesma pasta

    print(f"Attempting to convert {notebook_file} to PDF...")

    # Este comando só funcionará se nbconvert e as dependências LaTeX estiverem instaladas corretamente,
    # o que é improvável no ambiente padrão do Colab para geração de PDF direto.
    # Se falhar, considere converter para HTML primeiro e depois usar outra ferramenta.
    # !jupyter nbconvert --to pdf "{notebook_path}" --output "{output_path}"

    # Alternativa: converter para HTML (mais provável de funcionar)
    output_html_filename = os.path.splitext(notebook_file)[0] + '.html'
    output_html_path = os.path.join(notebooks_folder, output_html_filename)
    !jupyter nbconvert --to html "{notebook_path}" --output "{output_html_path}"
    print(f"Converted {notebook_file} to HTML at {output_html_path}")

    # Converter HTML para PDF é outro desafio no Colab e geralmente requer ferramentas
    # como `wkhtmltopdf` ou `chromium` headless, que são difíceis de instalar e usar aqui.
    # Uma opção seria baixar o HTML e convertê-lo localmente.

print("Conversion process finished. Check your folder for output files.")
```

```python
# prompt: faz o download de todos os arquivos .html gerados em um arquivo .zip

import shutil
from google.colab import files

# Listar todos os arquivos .html gerados
html_files = [f for f in os.listdir(notebooks_folder) if f.endswith('.html')]

# Nome do arquivo zip
zip_filename = 'html_notebooks.zip'
zip_filepath = os.path.join(notebooks_folder, zip_filename)

# Criar o arquivo zip contendo os arquivos .html
# Note: make_archive espera o nome do arquivo (sem extensão .zip),
# o formato ('zip'), e o diretório base onde estão os arquivos.
# Se quisermos incluir apenas arquivos específicos, podemos usar a biblioteca zipfile.
# A abordagem com make_archive é mais simples se todos os arquivos a serem
# zipados estão em um único diretório e queremos zipar o conteúdo desse diretório.
# Para zipar arquivos específicos, é melhor usar a biblioteca zipfile.

import zipfile

with zipfile.ZipFile(zip_filepath, 'w') as zipf:
    for file in html_files:
        file_path = os.path.join(notebooks_folder, file)
        zipf.write(file_path, os.path.basename(file_path)) # Adiciona o arquivo no zip com seu nome original

print(f"Created zip file: {zip_filepath}")

# Fazer o download do arquivo zip
files.download(zip_filepath)
print(f"Downloaded zip file: {zip_filepath}")
```
