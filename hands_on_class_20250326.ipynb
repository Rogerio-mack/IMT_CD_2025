# **Índices, Localizadores `.loc` e `.iloc`**


## Sumário

* Seleções (`read-only`), empregue

>> `df[ row-selection-criteria ][ columns-selection-list ]`

* Alterações (`update`) por chave (rótulo) do índice ou por posição, empregue

>> `df.loc[ rows-key, column-key (name) ]...`, ou

>> `df.iloc[ rows-position, column-position ]...`

* Alterações (`update`) com seleção de dados, empregue por chave (rótulo) do índice ou por posição, empregue

>> `df.loc[ row-selection-criteria, column-key (name) ]...`,  ou

>> `df.iloc[ row-selection-criteria.index, column-position ]...`

* Query

>> `df.query[ ' non-predfix row-selection-criteria  ' ]`, como `df.query['blood="AB"']`

# Seleção de dados

Alternativamente a

`df[ row-selection-criteria ][ columns-selection-list ]`

você pode fazer,

`df.loc[ row-selection-criteria, column-key (name) ]` ou

`df.iloc[ row-selection-criteria.index, column-position ]`

```python
df[ df.cyl == 8 ][['mpg','cyl','hp']].head()
```

```python
df.loc[ df.cyl == 8 , ['mpg','cyl','hp'] ].head()
```

```python
# Embora funcione, não é uma boa solução:
df.loc[ df.cyl == 8 ][ ['mpg','cyl','hp'] ].head()
```

# Atualizando valores

Seleções de dados do modo

`df[ row-selection-criteria ][ columns-selection-list ]`

criam uma *visão dos dados*, e não permitem alterações. Por isso, alterações devem sempre ser feitas com os operadores de localização `.loc` ou `.iloc`.

```python
df = pd.read_csv('https://vincentarelbundock.github.io/Rdatasets/csv/datasets/mtcars.csv',index_col='rownames')
df.index.name = ''
df.head()
```

## Novas colunas e séries completas

Novas colunas ou alterações de todas as linhas podem ser feitas diretamente, pois não envolvem uma seleção dos dados.

1.   Item da lista
2.   Item da lista



```python
df['kmpg'] = 1.6 * df['mpg']
df['hp'] = df['hp'] * 10
df['hp/mpg'] = df['hp'] / df['mpg']
df.head()
```

## `df.copy()`

```python
df2 = df
df2.loc['Hornet 4 Drive','mpg'] = 99
df2.head()
```

```python
id(df), id(df2)
```

```python
display(df.loc['Hornet 4 Drive','mpg'])
df2.loc['Hornet 4 Drive','mpg'] = 999
display(df.loc['Hornet 4 Drive','mpg'])
display(df2.loc['Hornet 4 Drive','mpg'])
```

```python
df2 = df.copy()
df2.loc['Hornet 4 Drive','mpg'] = 0
df2.head()
```

## Falhas nas alterações com views

```python
df[(df.gear > 0)]['mpg'] = 99
df.head()
```

```python
df[(df.gear == 4)]['mpg'] = 99
df.head()
```

```python
df.loc['Datsun 710':'Ford Pantera L']['hp'] = 999
df.head()
```

## Alterações corretas

```python
df.loc['Hornet 4 Drive','mpg'] = -1
df.head()
```

```python
df.iloc[3,0] = -100
df.head()
```

```python
df.loc[df.cyl == 6,'mpg'] = -1
df.head()
```

# Iterando sobre o dataframe

## `iloc`

```python
for i in range(0,len(df)):
  for j in range(0,len(df.columns)):
    if df.iloc[i,j] < 0:
      print( df.iloc[i,j] )
```

## `.loc`

```python
df.keys()
```

```python
for line in df.index:
  for col in df:
    if df.loc[line,col] < 0:
      print( df.loc[line,col] )
```

## `iterrows()`

```python
for index, row in df.iterrows():

  if row['disp'] > 450:
    print(row)
```

## items



```python
for col, data in df.items():
  print(col, data)
```
