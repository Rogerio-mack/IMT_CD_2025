# **Groupby**

```python
import pandas as pd

df = pd.read_csv('https://vincentarelbundock.github.io/Rdatasets/csv/MASS/Cars93.csv',index_col=0)
df.index.name = ''
df.head()
```

# Group by simples

```python
df.groupby('Type').Price.mean()
```

```python
df.groupby('Type')
```

```python
df.groupby('Type')[['Price','MPG.city']].mean()
```

```python
df.groupby(['Type','DriveTrain'])[['Price','MPG.city']].mean()
```

Isso é um dataframe, mas cuidado com os índices compostos.

```python
df.groupby(['Type','DriveTrain'])[['Price','MPG.city']].mean().reset_index()
```

```python
df.groupby('Type').Price.agg(['mean','std'])
```

```python
df.groupby(['Type','DriveTrain'])[['Price','MPG.city']].agg({ 'Price': ['mean', 'std'], 'MPG.city': 'max' })
```
